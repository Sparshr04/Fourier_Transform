### Another Example with fft methods

```python
# Function Declaration
def fourier_transform(input, sampling):
    time = np.linspace(0, 1, sampling, endpoint=False)
    signal = input(time)

    # Fourier transform
    fourier_transformed = np.fft.fft(signal)
    frequency_array = np.fft.fftfreq(len(signal), 1/sampling)

    return frequency_array, fourier_transformed
```

```python
# Example of a distorted sine wave:
def sine(t):
  x = np.sin(2*np.pi*5*t) + 0.3*np.sin(2*np.pi*20*t)
  return x
```

```python
# Calling Functions
sampling = int(input("Enter the Sampling Rate (defines Smoothness): "))
print('\nThe Input Array is: ')
frequency, spectrum = fourier_transform(sine, sampling)
print(frequency)
```

```python
# Plotting the Input Function
plt.plot(np.linspace(0, 1, sampling), sine(np.linspace(0, 1, sampling)))
plt.title("Sine Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Plotting the Fourier Transform
plt.plot(frequency, np.abs(spectrum))
plt.title("Fourier Transform of a Sine Wave")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(-300,300)
plt.grid(True)
plt.show()
```
